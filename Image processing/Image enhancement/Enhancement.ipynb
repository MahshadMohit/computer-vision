**Image enhancement**
 is the process of improving the quality and appearance of an image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow

image = cv2.imread("/content/dog-smiling.png")
```

**Adjusting brightness and contrast**

g(x)=α⋅f(x)+β

α -> contrast (mult pix)

β -> brightness (add pix)

```python
plt.subplot(1, 2, 1)
plt.title("Original Dog")
plt.imshow(image)

img = cv2.addWeighted(image , 2 , np.zeros(image.shape, image.dtype), 0, 10)

plt.subplot(1, 2, 2)
plt.title("Contrasted Dog")
plt.imshow(img)

cv2_imshow(cv2.resize(img,(200,200),interpolation=cv2.INTER_AREA))
```

**Sharpening**

enhancing the edges and fine details


1. Kernel

```python

kernel = np.array([
    [0, -1, 0],
    [-1, 5.7, -1],
    [0, -1, 0]
])

img = cv2.filter2D(image,-1,kernel)

cv2_imshow(cv2.resize(image,(200,200),interpolation=cv2.INTER_AREA))

cv2_imshow(cv2.resize(img,(200,200),interpolation=cv2.INTER_AREA))

cv2_imshow(cv2.resize(cv2.Laplacian(image, cv2.CV_64F),(200,200),interpolation=cv2.INTER_AREA))
```

**Laplacian**

```python
plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(cv2.resize(image,(200,200),interpolation=cv2.INTER_AREA))

plt.subplot(1,3,2)
plt.title("N1")
plt.imshow(cv2.resize(img,(200,200),interpolation=cv2.INTER_AREA))

plt.subplot(1,3,3)
plt.title("N2")
plt.imshow(cv2.resize(cv2.Laplacian(image, cv2.CV_64F),(200,200),interpolation=cv2.INTER_AREA))
```

**Removing noise**

1. median
2. Gaussian

```python
img = cv2.medianBlur(image,15)
cv2_imshow(cv2.resize(image,(200,200),interpolation=cv2.INTER_AREA))
cv2_imshow(cv2.resize(img,(200,200),interpolation=cv2.INTER_AREA))

img = cv2.GaussianBlur(image, (7,7), 0)
cv2_imshow(cv2.resize(img,(200,200),interpolation=cv2.INTER_AREA))
```

**Inverse image**

```python
img = 255 - image

cv2_imshow(cv2.resize(image,(200,200),interpolation=cv2.INTER_AREA))
cv2_imshow(cv2.resize(img,(200,200),interpolation=cv2.INTER_AREA))
```
