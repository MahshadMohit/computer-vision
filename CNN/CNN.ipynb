Layers:

Input: raw input of the image or data

Convolution:
extract features from input

applies kernels to the input and smaller metrics

it computes dot product between kernel weight and input image patch

the output would be a feature map


Activation layer:

they add nonlinearity to network:

RELU - Tanh - Leaky RELU

Pooling layer:

reduce size of volume

1. max pooling
2. average pooling


Flattening:

one-dimensional vector

Fully connected layer:

final classification or regression


output layer:

sigmoid or softmax function to probability
